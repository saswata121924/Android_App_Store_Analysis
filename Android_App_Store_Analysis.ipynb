```python
import pandas as pd
df_apps = pd.read_csv("apps.csv")
print(df_apps.shape)
print(df_apps.columns)
df_apps.head()
```

```python
df_apps.drop(["Last_Updated", "Android_Ver"], axis =1, inplace=True)
df_apps.head()
```

```python
print(df_apps["Rating"].isna().values.any())
df_apps[df_apps["Rating"].isna()].count()
```

```python
df_apps_clean = df_apps.dropna(subset=["Rating"])
```

```python
print(df_apps_clean.shape)
df_apps_clean.head()
```

```python
df_apps_clean[df_apps_clean.duplicated()].head()
```

```python
df_apps_clean.drop_duplicates(subset=["App", "Type", "Price"], inplace=True)
df_apps_clean.head()
```

```python
df_apps_clean.sort_values(by=["Rating"], ascending=[False], inplace=True)
df_apps_clean.head(10)
```

```python
df_apps_clean.sort_values(by="Size_MBs", ascending=False, inplace=True)
df_apps_clean.head()
```

```python
df_apps_clean.sort_values(by="Reviews", ascending=False, inplace=True)
df_apps_clean.head(50)
```

```python
import plotly.express as px
ratings = df_apps_clean["Content_Rating"].value_counts()
ratings
```

```python
px.pie(labels=ratings.index,
       values=ratings.values,
       names=ratings.index,
       title="Content Rating"
       ).update_traces(textposition="outside", textinfo="percent")
```

```python
df_apps_clean["Installs"].describe()
```

```python
df_apps_clean["Installs"] = df_apps_clean["Installs"].astype(str).str.replace(",","")
df_apps_clean["Installs"] = pd.to_numeric(df_apps_clean["Installs"])
df_apps_clean[["App", "Installs"]].groupby("Installs").count()
```

```python
df_apps_clean["Price"] = df_apps_clean["Price"].astype(str).str.replace("$","")
df_apps_clean["Price"] = pd.to_numeric(df_apps_clean["Price"])
```

```python
df_apps_clean.sort_values("Price", ascending=False).head(20)
```

```python
df_apps_clean = df_apps_clean[df_apps_clean["Price"]<299]
df_apps_clean.sort_values(by="Price", ascending=False).head(10)
```

```python
df_apps_clean["Revenue Estimate"] = df_apps_clean["Price"].mul(df_apps_clean["Installs"])
df_apps_clean.sort_values(by="Revenue Estimate", ascending=False).head(10)
```

```python
df_apps_clean["Category"].nunique()
```

```python
category_counts = df_apps_clean["Category"].value_counts()
category_counts.head(10)
```

```python
px.bar(x=category_counts.index, y=category_counts.values, title="Category Popularity")
```

```python
category_conc = pd.DataFrame({"App": df_apps_clean.groupby("Category")["App"].count(),
                              "Installs": df_apps_clean.groupby("Category")["Installs"].sum()})
category_conc.sort_values(by="Installs", ascending=False).head(10)
```

```python
px.scatter(category_conc,
           x="App",
           y="Installs",
           title="Category Concentration",
           size='App',
           hover_name=category_conc.index,
           color="Installs"
           ).update_layout(xaxis_title="Number of Apps (Lower=More Concentrated)",
                           yaxis_title="Installs",
                           yaxis=dict(type='log'))
```

```python
df_apps_clean.Type.value_counts()
```

```python
df_free_vs_paid = df_apps_clean.groupby(['Category', 'Type'], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.head()
```

```python
px.bar(df_free_vs_paid,
       x="Category",
       y="App",
       color='Type',
       hover_name = df_free_vs_paid['Type'],
       title='Free vs Paid Apps by category',
       barmode='group').update_layout(xaxis_title="Category",
                                      yaxis_title="Number of Apps",
                                      xaxis={'categoryorder': 'total descending'},
                                      yaxis = dict(type='log'))
```

```python
px.box(df_apps_clean,
       x="Type",
       y="Installs",
       notched=True,
       points='all',
       title="How many Downloads are Paid Apps Giving Up?").update_layout(xaxis_title="Type",
                                                                          yaxis_title="Installs",
                                                                          yaxis=dict(type='log'))
```

```python
df_paid_apps = df_apps_clean[df_apps_clean["Type"]=="Paid"]
px.box(df_paid_apps,
       x="Category",
       y="Revenue Estimate",
       title="How much can paid Apps earn?").update_layout(xaxis_title="Category",
                                                           yaxis_title="Paid App Ballpark Revenue",
                                                           yaxis=dict(type='log'),
                                                           xaxis={'categoryorder': 'min ascending'})
```
